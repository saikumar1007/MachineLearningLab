```python
#Import the Necessary Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Load the data sets- train and test 
df = pd.read_csv('train_titanic.csv')
print(df)
X = df.drop(columns=['Survived']) # Features (independent variables)
y = df['Survived'] # Target (dependent variable)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

print("Training set size:", X_train.shape, y_train.shape)
print("Testing set size:", X_test.shape, y_test.shape)
```

```python
#Look at top values of train data 
display(X_train.head(10))  # Features
display(y_train.head(10))  # Target values (Survived or not)
```

```python
#Print the column names of train data
display(X_train.columns) # Train Features 
display(y_train.name) # Train Target 
```

```python
#Print the dimension of train and test data
print("dimensions of the train data : ", X_train.shape, y_train.shape)
print("dimensions of the train data : ", X_test.shape, y_test.shape)
```

```python
#Print the number of missing values in each column of train and test data
print("missing values in the features and the treger in the train data respectively: ",X_train.isnull().sum(),"\nTarget: ",y_train.isnull().sum())
print("missing values in the features and the treger in the test data respectively: ",X_test.isnull().sum(),"\nTarget: ",y_test.isnull().sum())
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Barchart of Survived, Pclass, Sex, Embarked

# Set Seaborn style
sns.set_theme(style="darkgrid")

# List of categorical columns to plot
columns = ["Survived", "Pclass", "Sex", "Embarked"]

# Create bar charts for each column
for col in columns:
  plt.figure(figsize=(3,2))
  sns.countplot(x=df[col],palette="bright")  # Use countplot for categorical data
  plt.title(f"Distribution of {col}")
  plt.show()

```

```python
#Histogram of Age, Parch, Fare
sns.set_theme(style="darkgrid")
columns = ["Age", "Parch", "Fare"]

for col in columns:
  plt.figure(figsize=(10,2))
  sns.histplot(data = df[col],bins=10,kde=True)
  plt.title(f"Distribution of{col}")
  plt.xlabel(col)
  plt.ylabel("Freqeuncy")
  plt.show()
```

```python
#Plot histogram of age for survived and did not survived class
plt.figure(figsize=(10, 3))
sns.histplot(data=df, x="Age", bins=20, hue="Survived", palette=["red", "green"], kde=True)
plt.show()
```

```python
#Count class and geder wise survival rate - plot with appropriate chart
plt.figure(figsize=(10, 5))
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Class & Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#Age wise Distribution of Male and Female passengers
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="Age", hue="Sex", bins=20, kde=True, palette=["blue", "pink"])

plt.title("Age-wise Distribution of Male & Female Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Histogram of Age of the given data set(sample)
plt.figure(figsize=(8, 4))
sns.histplot(data=df["Age"], bins=20, kde=True, color="skyblue")

# Labels and title
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
```
